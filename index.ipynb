## Create another column 'Num_Lines' that is a count of how many lines pass through a station. Then sort your dataframe by this column in descending order.

## Compare subway traffic by day of the week. Display this as a graph.

```python
#Your code here
traffic_dict = {}
df = df.sort_values(by=['C/A','Unit','Scp','Date','Time'])

scp = df.iloc[0]['Scp']
date = df.iloc[0]['Date']
day = df.iloc[0]['Day']
old_entries = df.iloc[0]['Entries']
new_entries = None

for i in range(1,127):
    
    if df.iloc[i]['Date'] == date:
        continue
    elif df.iloc[i]['Date'] != date and df.iloc[i]['Scp'] == scp:
        new_entries = df.iloc[i]['Entries']
        traffic_for_the_day = new_entries - old_entries
        traffic_dict[day] = traffic_dict.get(day,0) + traffic_for_the_day
                
        date = df.iloc[i]['Date']
        old_entries = new_entries
        day = df.iloc[i]['Day']
    else:
        new_entries = df.iloc[i-1]['Entries']
        traffic_for_the_day = new_entries - old_entries
        traffic_dict[day] = traffic_dict.get(day,0) + traffic_for_the_day
        
        date = df.iloc[i]['Date']
        old_entries = df.iloc[i]['Entries']
        day = df.iloc[i]['Day']

import matplotlib.pyplot as plt
days = []
traffic_counts = []

for key in traffic_dict.keys():
    days.append(key)
    traffic_counts.append(traffic_dict[key])
    
plt.bar(days, traffic_counts, label="Subway Traffic from Three turnstiles")
plt.xlabel("Day of the Week")
plt.ylabel("Traffic Counts")
plt.title("Subway Traffic Counts by Day of the Week")
plt.legend()
plt.show()
```
